# Linear Regression

```python
# predicting life expectancy based on spending USD--
df=data[['Spending_USD','Life_Expectancy']]
df.head()
```

```python
df.shape
```

```python
# sampling of data--
training,testing=train_test_split(df,test_size=0.3)
print(training.shape)
print(testing.shape)
```

```python
x = training['Spending_USD'].values.reshape(-1,1)
y = training['Life_Expectancy'].values.reshape(-1,1)
```

```python
model=LinearRegression()
print(model)
```

```python
training_model=model.fit(x,y)
print("Model Training is done")
print(training_model)
```

```python
print(model.coef_)        ### this is m value
print(model.intercept_)   ### this is b value 
                          ### y = mx + b
```

```python
### Equation after training the model
x = 10000
y = round((model.coef_[0][0])*x + (model.intercept_[0]))
print("The Life Expectancy of Spending USD {} is {} years.".format(x,y))
```

```python
x_test=testing.Spending_USD.values.reshape(-1,1)
x_test
```

```python
pred_LifeExp=training_model.predict(x_test)
pred_LifeExp
```

```python
testing['pred_LifeExp'] = pred_LifeExp
```

```python
testing
```

```python
sns.regplot(x = 'Spending_USD', y = 'Life_Expectancy', data = training)
```

```python
###  Inference - 1. Life Expectancy is highest i.e. 85 years with Spending USD being approx. 5000.
###              2. Spending USD is highest i.e. 12000 with Life Expectancy approx. 76 years.
```

```python
sns.regplot(x = 'Spending_USD', y = 'pred_LifeExp', data = testing)
```

```python
###  Inference - Predicted Life Expectancy increases constantly with increasing Life Expectancy.
```
